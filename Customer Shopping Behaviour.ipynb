NOW I WANT TO SEE A LINE CHART SHOWING THE NUMBER OF PURCHASES BY CATEGORY

```python
category_counts= df['CATEGORY'].value_counts()
category_counts.plot(kind='line', marker='o', title='NUMBER OF PURCHASES BY CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('NUMBER OF PURCHASES')
plt.grid()
plt.xticks(rotation=20)
plt.show()
```

NEXT LETS TAKE LOOK AT TOP 10 MOST PURCHASES ITEMS IN BAR CHART

```python
df['ITEM_PURCHASED'].value_counts().head(10).plot(kind='bar', figsize=(10,6))
plt.title("Top 10 Most Purchased Items")
plt.xlabel("ITEM")
plt.ylabel("COUNT")
plt.xticks(rotation=40) 
plt.tight_layout() 
plt.grid(axis='y', linestyle='--') 
plt.show()
```

NEXT LETS TAKE LOOK AT DISTRIBUTION OF PAYMENT METHODS IN PIE CHART

```python
df['PAYMENT_METHOD'].value_counts().plot(kind='pie', autopct='%1.1f%%',startangle=130, figsize=(6,6))
plt.title("PAYMENT METHOD USAGE")
plt.ylabel('')
plt.show()
```

NEXT LETS TAKE LOOK AT NUMBER OF PURCHASES ACROSS SEASONS IN BAR CHART

```python
sns.countplot(x='SEASON', data=df, order=df['SEASON'].value_counts().index)
plt.title("PURCHASES BY SEASON")
plt.xlabel("SEASON")
plt.ylabel("NUMBER OF PURCHASES")
plt.grid(axis='y', linestyle='--')
plt.show()
```

NEXT LETS TAKE LOOK AT GENDER-WISE PURCHASES BY BREAKDOWN IN PIE CHART

```python
plt.figure(figsize=(8, 6)) 
counts = df['GENDER'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title('PURCHASE BY GENDER')
plt.legend(title='GENDER')
plt.show()
```

NEXT LETS TAKE LOOK AT PURCHASES BY GENDER WITHIN EACH CATEGORY IN STACKED BAR CHART

```python
cross_tab = pd.crosstab(df['CATEGORY'], df['GENDER'])
cross_tab.plot(kind='bar', stacked=True, figsize=(8, 6))

plt.title('CATEGORY BY GENDER')
plt.xlabel('CATEGORY')
plt.ylabel('COUNT')
plt.xticks(rotation=25)
plt.legend(title='GENDER')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

NEXT LETS TAKE LOOK AT CUSTOMER SEGMENTATION BASED ON AGE IN HORIZONTAL BAR CHART

```python
age_counts = df['AGE'].value_counts().sort_index().head(10)
plt.figure(figsize=(8, 6))
age_counts.plot(kind='barh', color='brown')
plt.xlabel('NUMBER OF CUSTOMERS')
plt.ylabel('AGE GROUP')
plt.title('MOST NUMBER OF CUSTOMERS BY AGE GROUP')
plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()
```
